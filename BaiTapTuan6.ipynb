```python
#W6A4
s = map(str, input().split())
ans = {}
for p in s:
    if ':' not in p: 
        continue
    key,val = p.split(':',1)
   # val = int(val)
    if key in ans:
        ans[key].append(val)
    else: 
        ans[key]=[val]
    
print(ans)
#a:1 a:2 b:2 c:3 c:4 a:5
```

```python
#W6A5
num = list(map(int,input().split()))
ans = {
    "positives":0,
    "negatives":0,
    "zeros":0
}
for x in num:
    if x > 0: ans["positives"]+=1
    elif x < 0: ans["negatives"]+=1
    else: ans["zeros"]+=1
print(ans)
# -5 6 7 0 0 8 9 -2 0
#  4 2 3
```

```python
#W6A9
s1 = list(map(str,input().split()))
s2 = list(map(str,input().split()))
d1 = {}
d2 = {}
ans={}
for x in s1:
    key,val = x.split(':',1)
    d1[key]=int(val)
for x in s2:
    key,val = x.split(':',1)
    d2[key]=int(val)
for k,v in d1.items():
    ans[k]=v
for k,v in d2.items():
    if k in ans:
        ans[k]+=v
    else:
        ans[k]=v 
sorted(ans)
print(ans)
#a:1 b:2 c:3 b:5 d:4
```

```python
#W6A10
num = list(map(int,input().split()))
k = int(input())
n=len(num)
ans=set()
for i in num:
    for j in num:
        if num[i]+num[j]==k and i < j:
            ans.add((num[i],num[j]))
ans=list(ans)
ans=sorted(ans)
print(ans)
# 3 1 4 2 5 2 k=6 
```

```python
#W6A14
num1 = list(map(int,input().split()))
num2 = list(map(int,input().split()))
ans = set()
for x in num1:
    for y in num2:
        if x == y:
            ans.add(x)
ans = list(ans)
print(ans)
# 1 2 3 4 10 10 12 # 2 5 9 4 10 5 12 10
```
