**Bài 2:** iết chương trình máy tính:
- Nhập 2 số a, b
- Nhập phép toán (+, -, *, /)
- In kết quả tương ứng
- Với phép chia, kiểm tra b != 0
- Nếu phép toán không hợp lệ, in "Phép toán không hợp lệ"

Ví dụ:
Nhập số thứ nhất: 10
Nhập số thứ hai: 5
Nhập phép toán (+, -, *, /): +
Kết quả: 10 + 5 = 15

```python
a = float(input("Nhập số thứ nhất: "))
b = float(input("Nhập số thứ hai: "))
o = input("Nhập phép toán (+, -, *, /): ")
# Viết code ở đây
if o == "+":
    print("a+b = ", a+b)
if o == "-":
    print("a-b = ", a-b)
if o == "*":
    print("a*b = ", a*b)
if o == "/":
    if b!=0:
        print("a/b = ", a/b)
    else:
        print("Phép toán không hợp lệ")
```

**Bài 6:** Viết chương trình tính tiền điện theo bậc thang:
- Bậc 1: 0-50 kWh: 1,678đ/kWh
- Bậc 2: 51-100 kWh: 1,734đ/kWh
- Bậc 3: 101-200 kWh: 2,014đ/kWh
- Bậc 4: 201-300 kWh: 2,536đ/kWh
- Bậc 5: 301-400 kWh: 2,834đ/kWh
- Bậc 6: Trên 400 kWh: 2,927đ/kWh

Input:
- Chỉ số cũ
- Chỉ số mới

Output:
- Điện tiêu thụ: ... kWh
- Tiền bậc 1: ...
- Tiền bậc 2: ...
- ...
- Tổng tiền (chưa VAT): ...
- VAT (10%): ...
- Tổng thanh toán: ...

```python
old_index = int(input("Chỉ số cũ: "))
new_index = int(input("Chỉ số mới: "))
# Viết code ở đây
tieu = new_index - old_index
rates = [1678, 1734, 2014, 2536, 2834, 2927]
costs = [0, 0, 0, 0, 0, 0]
x = tieu
if x > 0:
    used = min(x, 50)
    costs[0] = used * rates[0]
    x -= used
if x > 0:
    used = min(x, 50)
    costs[1] = used * rates[1]
    x -= used
if x > 0:
    used = min(x, 100)
    costs[2] = used * rates[2]
    x -= used
if x > 0:
    used = min(x, 100)
    costs[3] = used * rates[3]
    x-= used
if x > 0:
    used = min(x, 100)
    costs[4] = used * rates[4]
    x -= used
if x > 0:
    costs[5] = x * rates[5]
total = sum(costs)
vat = total * 0.1
print("Điện tiêu thụ:", tieu, "kWh")
for i in range(6):
    print(f"Tiền bậc {i+1}: {costs[i]}")
print("Tổng tiền (chưa VAT):", total)
print("VAT (10%):", vat)
print("Tổng thanh toán:", total + vat, "đ")
```

**Bài 7:** Viết chương trình đặt vé xem phim:

Input:
- Tuổi khách hàng
- Loại ghế (Standard/VIP/Couple)
- Ngày xem (Weekday/Weekend)
- Suất chiếu (Morning/Afternoon/Evening)
- Có thẻ thành viên không (y/n)

Giá vé cơ bản:
- Standard: 45,000đ
- VIP: 70,000đ
- Couple: 150,000đ

Điều chỉnh giá:
1. Theo tuổi:
   - Dưới 6 tuổi: Miễn phí
   - 6-12 tuổi: -30%
   - Trên 60 tuổi: -20%

2. Theo ngày:
   - Weekend: +20%

3. Theo suất chiếu:
   - Morning (trước 12h): -20%
   - Evening (sau 18h): +10%

4. Thẻ thành viên: -15%

In ra hóa đơn chi tiết

```python
age = int(input("Tuổi: "))
seat_type = input("Loại ghế (Standard/VIP/Couple): ")
day_type = input("Ngày xem (Weekday/Weekend): ")
show_time = input("Suất chiếu (Morning/Afternoon/Evening): ")
member = input("Có thẻ thành viên? (y/n): ")
# Viết code ở đây
loaighe = { "Standard": 45000, "VIP": 70000, "Couple": 150000 }
price = loaighe.get(seat_type) # Giá vé ban đầu theo loại ghế

print("Tuổi khách hàng: ", age)
print("Giá vé: ", price)
if age < 6:
    price = 0
    print("Dưới 6 tuổi: Miễn phí")
elif 6 <= age <= 12:
    discount = price * 0.3
    price -= discount
    print("Giảm giá 30% cho trẻ em:", "-", discount)
elif age > 60:
    discount = price * 0.2
    price -= discount
    print("Giảm giá 20% cho người trên 60 tuổi:", "-", discount)
if day_type == "Weekend":
    tangcuoituan = price * 0.2
    price += tangcuoituan
    print("Cuối tuần +20%:", "+", tangcuoituan)

if show_time == "Morning":
    discount = price * 0.2
    price -= discount
    print("Suất sáng -20%:", "-", discount)
elif show_time == "Evening":
    increase = price * 0.1
    price += increase
    print("Suất tối +10%:", "+", increase)

if member == "y":
    discount = price * 0.15
    price -= discount
    print("Giảm giá thành viên -15%:", "-", discount)
print("TỔNG THANH TOÁN:", price)
```

**Bài 8:**  Viết chương trình tính lương:

Input:
- Số giờ làm việc
- Lương theo giờ
- Loại ngày (weekday/weekend/holiday)
- Đánh giá (poor/average/good/excellent)

Quy tắc:
1. Tính lương theo loại ngày:
   - Weekday:
     * 8 giờ đầu: x1.0
     * Giờ 9-10: x1.5
     * Từ giờ 11: x2.0
   - Weekend: toàn bộ x1.5
   - Holiday: toàn bộ x2.0

2. Thưởng/phạt hiệu suất:
   - poor: -10%
   - average: 0%
   - good: +10%
   - excellent: +20%

3. Thưởng chuyên cần:
   - Làm >= 48 giờ: +500,000đ

Output chi tiết:
- Lương giờ hành chính: ...
- Lương tăng ca: ...
- Tổng lương cơ bản: ...
- Điều chỉnh hiệu suất: ...
- Thưởng chuyên cần: ...
- Tổng lương: ...

```python
x = float(input("Số giờ làm việc: "))
hourly_rate = float(input("Lương theo giờ: "))
day_type = input("Loại ngày (weekday/weekend/holiday): ")
performance = input("Đánh giá (poor/average/good/excellent): ")
luonggiohanhchinh = 0
luongtangca = 0
dchs = 0
chuyencan = 0
if day_type == 'weekday':
    if x <= 8:
        luonggiohanhchinh = hourly_rate * x
    elif x <= 10: 
        luonggiohanhchinh = 8 * hourly_rate
        luongtangca = (x - 8) * hourly_rate * 1.5 
    else:  
        luonggiohanhchinh = 8 * hourly_rate
        luongtangca = 2 * hourly_rate * 1.5 + (x - 10) * hourly_rate * 2.0
elif day_type == 'weekend': 
    luongtangca = hourly_rate * 1.5 * x
elif day_type == 'holiday':
    luongtangca = hourly_rate * 2.0 * x

tongluongcoban = luonggiohanhchinh + luongtangca

if performance == 'poor':
    dchs = tongluongcoban * (-0.1)
elif performance == 'average':
    dchs = 0
elif performance == 'good':
    dchs = tongluongcoban * 0.1
elif performance == 'excellent':
    dchs = tongluongcoban * 0.2

if x >= 48:
    chuyencan = 500000

tongluong = tongluongcoban + dchs + chuyencan

print(f"Lương giờ hành chính: {luonggiohanhchinh} ")
print(f"Lương tăng ca: {luongtangca} ")
print(f"Tổng lương cơ bản: {tongluongcoban}")
print(f"Điều chỉnh hiệu suất: {dchs} ")
print(f"Thưởng chuyên cần: {chuyencan}")
print(f"Tổng lương: {tongluong}")
```

Bài 9: Chương trình sau có lỗi. Hãy:
1. Thêm các dòng print để trace control flow

3. Giải thích lỗi là gì?

```python
price = float(input("Giá sản phẩm: "))
quantity = int(input("Số lượng: "))
customer_type = input("Loại khách (VIP/Regular): ")

discount = 0

if customer_type == "VIP":
    if quantity > 10:
        discount = 0.2
    else:
        discount = 0.1
elif customer_type == "Regular":
    if quantity > 20:
        discount = 0.15
    else:
        discount = 0.05

# Lỗi ở đây!
final_price = price * quantity * (1 - discount)
print(f"giảm giá: {discount*100}%")
print(f"Giá cuối: {final_price}")
```
