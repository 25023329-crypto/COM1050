```python
#W5A3
def snt(x):
    if x < 2 : return False 
    else: 
        for i in range (2, int(x**0.5)+1):
            if x % i == 0: return False 
    return True 
n = int(input())
if snt(n): print(n, "là snt")
else: print(n, "Kp snt")
```

```python
#W5A4
def tonguoc(x):
    s = 0
    if x == 1:
        return 0 
    
    s = 1 
    for i in range(2, int(x**0.5) + 1): 
        if x % i == 0:
            s += i
            j = x // i
            if j != i: 
                s += j
    return s

n = int(input())
if tonguoc(n) == n:
    print("Y")
else:
    print("N")
```

```python
#W5A4
def tong(n):
    s=1 #tổng các ước
    temp = n
    i=2
    while i*i<=temp:
        if temp%i==0: #nếu i là thừa số ngto
            power_sum=1  #tổng 1+i+i^2+...
            p=1          #i^0=1 ban đầu
            while temp%i == 0: #đếm bội số của i
                temp //= i 
                p *=i
                power_sum += p
            s*=power_sum 
        i+=1
    if temp > 1:
        s*= (1+temp)
    return s 
n = int(input())
if n == tong(n) - n:
    print("hoàn hảo")
print("không hoàn hảo")
```

```python
#W5A9 
def tong(n):
    sum=0
    while n >= 10:
        sum += n%10
        n//=10
    sum+=n 
    return sum 
n = int(input())
print(tong(n))
```

```python
#W5A10
def mapping_check(x,y):
    if len(x) != len(y): return False 
    mp={}
    used=set()
    for a,b in zip(x,y):
        if a in mp:
            if mp[a]!=b: return False 
        else:
            if b in used:
                return False 
            else: 
                mp[a]=b
                used.add(b)
    return True 
a,b = map(str, input().split())
print(mapping_check(a,b))
```
