```python
#W4A8
def reverse(n):
    return int(str(n)[::-1])
def palindrome(n):
    x = str(n)
    return x == x[::-1]
n = int(input())
count=0
if n < 0: print("nhap lai")
if n > 0:
    while not palindrome(n):
        n = n + reverse(n)
        count+=1
    print("So buoc: ",count)
    print("Gia tri palindrome: ",n)

```

```python
#W4A9
def scp(n):
    if n < 0:
        return False 
    
    x = int(n**0.5)
    return x*x == n 
def khac(n):
    x = str(n)
    return len(x) == len(set(x))
n = int(input())
found = False 
for i in range(0,n+1):
    if scp(i) and khac(i):
        print(i, end = ' ')
        found = True
if not found:
    print("no number")

    
```

```python
#W4A10
def longestcollatz(n):
    cnt = 1
    while n!=1:
        if n%2 == 0:
            n/=2
            cnt+=1
        else:
            n = n*3+1
            cnt +=1
    return cnt 
n = int(input())
ans = 0
maxx = 0
a = 0
for x in range(1, n+1):
    a = longestcollatz(x)
    if a > maxx: 
        maxx = a 
        ans = x
print(ans, maxx)
```

```python
#W4A13
def closepair(a,b):
    a1 = 0
    b1 = 0
    for i in range(1, int(a/2+1)): 
        if a%i==0: a1+= i 
    for i in range(1, int(b/2+1)):
        if b%i==0: b1+= i
    if a1 == b and b1 == a: return True 
    return True 
m,n = list(map(int,input().split()))
if closepair(m,n):print("True")
else: print("False")

```
